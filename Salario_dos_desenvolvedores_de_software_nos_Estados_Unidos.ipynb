# Lendos os dados

```python
import zipfile
zf = zipfile.ZipFile('archive.zip')
```

```python
# Todos os caminhos internos
zf.namelist()
```

```python
# Como o caminho dos dados é 'SofwareDeveloperIncomeExpensesperUSACity.csv', temos que
df = pd.read_csv(zf.open('SofwareDeveloperIncomeExpensesperUSACity.csv'),
                 #nrows=0,
                 sep=',',
                 encoding= 'ISO-8859-1')
df
```

```python
df.shape
```

```python
# As lable das colunas dos dados
display(df.columns.values,len(df.columns.values),df.dtypes)
```

```python
df.shape
```

```python
# Medidas descritivas
round(df[['Mean Software Developer Salary (unadjusted)','Mean Unadjusted Salary (all occupations)','Cost of Living Plus Rent avg']].describe(),2)
```

```python
df['Metro']
```

```python
df[['Metro','Mean Software Developer Salary (unadjusted)','Mean Unadjusted Salary (all occupations)']]
```

```python
df['diferenca_ocupacoes'] = df['Mean Software Developer Salary (unadjusted)'] - df['Mean Unadjusted Salary (all occupations)']
df['diferenca_ocupacoes']
```

```python
df['diferenca_Salario_Ocupacoes_Bool'] = df['diferenca_ocupacoes']>0
df['diferenca_Salario_Ocupacoes_Bool']
```

```python
df['diferenca_Salario_Ocupacoes_Bool'].value_counts(dropna=True)
```

```python
df[['City','Mean Software Developer Salary (adjusted)','Cost of Living Plus Rent avg','Local Purchasing Power avg']]
```

```python
df['gasto_mensal'] = df['Cost of Living Plus Rent avg'] + df['Local Purchasing Power avg']
df['gasto_mensal']
```

```python
df[['City','Mean Software Developer Salary (adjusted)','gasto_mensal']]
```

```python
df['gasto_anual'] = df['gasto_mensal']*12
df['gasto_anual']
```

```python
df[['City','Mean Software Developer Salary (adjusted)','gasto_anual']]
```

```python
df['diferenca_gastoAno_salario'] = df['Mean Software Developer Salary (adjusted)'] - df['gasto_anual']
df['diferenca_gastoAno_salario']
```

```python
df['cidades_menosGasto'] = df['diferenca_gastoAno_salario']>0
df['cidades_menosGasto']
```

```python
df['cidades_menosGasto'].value_counts(dropna=True)
```

```python
df[df['cidades_menosGasto'] == True]
```

```python
import seaborn as sns; sns.set();
import matplotlib.pyplot as plt;
import plotly.express as px
dados = df[df['cidades_menosGasto'] == True].groupby('City')[['diferenca_gastoAno_salario']].mean().reset_index()
dados
plt.bar(dados.City,dados.diferenca_gastoAno_salario, color ='maroon',
        width = 0.4)
plt.xlabel("Cidades")
plt.ylabel("Resto salario")
plt.title("Cidades dos Estados Unidos em que o salario do desenvolvedor é maior que o gasto anual")
plt.savefig('Cidades.jpg')
plt.show()
plt.close()
```
